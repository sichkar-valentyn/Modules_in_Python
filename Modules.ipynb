```python
import requests

# The simple GET request
r = requests.get('https://example.com')
print(r.text)  # Showing the answer from server

# Adding the parameters to the request
url = 'http://example.com'
par = {'key1': 'value1', 'key2': 'value2'}
r = requests.get(url, params = par)
print(r.url)

# 
```

```python
# Implementing the task
# Getting the information from a set of files
import requests
# Initial adress with folder
url = 'https://stepic.org/media/attachments/course67/3.6.3/'
# The name of the first file
next_url = '699991.txt'
print(url + next_url)

while (next_url[0] + next_url[1]) != 'We':    
    # Get request
    r = requests.get(url + next_url)
    next_url = r.text
    print(next_url)

```
